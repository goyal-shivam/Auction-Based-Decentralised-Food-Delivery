```python
import simpy
import random

GAS_STATION_SIZE = 2  # Number of refill points
TANK_SIZE = 40  # Size of car's fuel tank
# REFUEL_SPEED = 2  # Litres of fuel per second
REFUEL_SPEED = 8  # Litres of fuel per second
ARRIVAL_INTERVAL = 5  # Time between car arrivals
SIM_TIME = 100  # Simulation time in seconds

def car(env, name, gas_station, fuel_pump):
    print(f'\n{name} enters the gas station at {env.now:.2f}')
    with gas_station.request() as req:
        yield req

        print(f'{name} starts refueling at {env.now:.2f}')

        # ye wali line panga kar deti hai
        yield fuel_pump.get(TANK_SIZE)

        yield env.timeout(TANK_SIZE / REFUEL_SPEED)
        print(f'{name} finishes refueling at {env.now:.2f}')

    print(f'{name} exits the gas station at {env.now:.2f}')

def car_generator(env, gas_station, fuel_pump):
    count = 0
    while True:
        yield env.timeout(random.expovariate(1/ARRIVAL_INTERVAL))
        count += 1
        env.process(car(env, f'Car {count}', gas_station, fuel_pump))

# Create the SimPy environment
env = simpy.Environment()

# Create the gas station and fuel pumps
gas_station = simpy.Resource(env, capacity=GAS_STATION_SIZE)
fuel_pump = simpy.Container(env, init=GAS_STATION_SIZE*TANK_SIZE, capacity=GAS_STATION_SIZE*TANK_SIZE)

# Start the simulation
env.process(car_generator(env, gas_station, fuel_pump))
```

```python
env.run(until=SIM_TIME)
```

```python
import simpy

GAS_STATION_SIZE = 2  # Number of refill points
TANK_SIZE = 40  # Size of car's fuel tank
REFUEL_SPEED = 2  # Litres of fuel per second
# ARRIVAL_TIMES =  [1, 2, 2, 3, 4, 5, 6, 8, 9, 10] # Arrival times for each car
ARRIVAL_TIMES =  [1, 1, 0, 1, 1, 1, 1, 2, 1, 1] # Arrival times for each car
REFILL_TIMES = [1, 2, 1, 2, 1, 2, 1, 2, 1, 2]  # Refill time for each car
SIM_TIME = sum(ARRIVAL_TIMES) + max(REFILL_TIMES)  # Simulation time in seconds

def car(env, name, gas_station, fuel_pump, refill_time):
    print(f'\n{name} enters the gas station at {env.now:.2f}')
    with gas_station.request() as req:
        yield req

        print(f'{name} starts refueling at {env.now:.2f}')
        
        # yield fuel_pump.get(TANK_SIZE)
        
        yield env.timeout(refill_time)
        print(f'{name} finishes refueling at {env.now:.2f}')

    print(f'{name} exits the gas station at {env.now:.2f} -------------------------------------------')

def car_generator(env, gas_station, fuel_pump, arrival_times, refill_times):
    for i, arrival_time in enumerate(arrival_times):
        yield env.timeout(arrival_time)
        env.process(car(env, f'Car {i+1}', gas_station, fuel_pump, refill_times[i]))
        

# Create the SimPy environment
env = simpy.Environment()

# Create the gas station and fuel pumps
gas_station = simpy.Resource(env, capacity=GAS_STATION_SIZE)
fuel_pump = simpy.Container(env, init=GAS_STATION_SIZE*TANK_SIZE, capacity=GAS_STATION_SIZE*TANK_SIZE)

# Start the simulation
env.process(car_generator(env, gas_station, fuel_pump, ARRIVAL_TIMES, REFILL_TIMES))
```

```python
# %pip install simpy

import pandas as pd
data = pd.read_pickle('/content/drive/MyDrive/aepr/data_preprocessed.pkl')

import random
from geopy.distance import geodesic
import simpy

# Constants
NUM_CABS = 5
NUM_BOYS = NUM_CABS
SIM_TIME = 120
CITY_COORDS = [(31, 0), (31, 88), (-30, 0.0), (-30, 88)]  # Coordinates of the corners of the city
# BOOKINGS = [(5, (31, 50), (20, 40)), (10, (20, 10), (-10, 20)), (15, (10, 20), (5, 60)),
#             (20, (0, 0), (-20, -20)), (25, (-10, 50), (10, 70)), (30, (-20, 20), (-30, 50)),
#             (35, (-20, 10), (-30, 0)), (40, (-5, 10), (-15, 30)), (45, (20, 40), (0, 60)),
#             (50, (5, 30), (-5, 70)), (55, (10, 40), (-5, 50)), (60, (-5, 20), (15, 40)),
#             (65, (-15, 50), (-5, 30)), (70, (20, 60), (5, 20)), (75, (30, 70), (-10, 50)),
#             (80, (-20, 0), (10, 30)), (85, (0, 50), (20, 20)), (90, (-10, 0), (20, 50)),
#             (95, (10, 60), (0, 10)), (100, (-30, 20), (0, 0))]

BOOKINGS = [(5, (31, 50), (20, 40)), (10, (20, 10), (-10, 20)), (15, (10, 20), (5, 60)),
            (20, (0, 0), (-20, -20)), (25, (-10, 50), (10, 70)), (30, (-20, 20), (-30, 50)),
            (35, (-20, 10), (-30, 0)), (40, (-5, 10), (-15, 30)), (45, (20, 40), (0, 60)),
            (50, (5, 30), (-5, 70)), (55, (10, 40), (-5, 50)), (60, (-5, 20), (15, 40)),
            (65, (-15, 50), (-5, 30)), (70, (20, 60), (5, 20)), (75, (30, 70), (-10, 50)),
            (80, (-20, 0), (10, 30)), (85, (0, 50), (20, 20)), (90, (-10, 0), (20, 50)),
            (95, (10, 60), (0, 10)), (100, (-30, 20), (0, 0))]

BOOKINGS = [(5, booking[1], booking[2]) for booking in BOOKINGS]

MAX_LAT = max(data['Restaurant_latitude'].max(), data['Delivery_location_latitude'].max())
MIN_LAT = min(data['Restaurant_latitude'].min(), data['Delivery_location_latitude'].min())

MAX_LONG = max(data['Restaurant_longitude'].max(), data['Delivery_location_longitude'].max())
MIN_LONG = min(data['Restaurant_longitude'].min(), data['Delivery_location_longitude'].min())


NUM_BOYS_PER_RESTRO = 5
BIKE_SPEED = 30

# Classes
class Cab:
    def __init__(self, env, id):
        self.env = env
        self.id = id
        self.location = self.get_random_location()
        self.status = "available"
        self.end_time = 0
        
    def get_random_location(self):
        lat = random.randrange(10001)
        long = random.randrange(10001)
        lat = MIN_LAT + (MAX_LAT - MIN_LAT)*lat/10000
        long = MIN_LONG + (MAX_LONG - MIN_LONG)*long/10000
        return (lat,long)
        # GADBAD
    
    def move_to(self, dest):
        '''GADBAD put status as unavailable'''
        distance = geodesic(self.location, dest).km
        time = distance / BIKE_SPEED * 60  # 30 km/hr is the speed of the cab
        yield self.env.timeout(time)
        self.location = dest
        
class Customer:
    def __init__(self, env, id, pickup, dropoff):
        self.env = env
        self.id = id
        self.pickup = pickup
        self.dropoff = dropoff
        self.start_time = env.now
        self.wait_time = None
        self.cab = None
        self.real_wait_time = None
        
    def request_cab(self, cabs, arrival_time):
      print(f'request cab function called for Customer #{self.id} at time {self.env.now}')
      yield env.timeout(arrival_time)
      print(f'Customer #{self.id} actually arrives at time {self.env.now}')
      with cabs.request() as req:
            yield req
            cab = self.find_nearest_cab(cabs)
            if cab is not None:
                self.cab = cab
                cab.status = "occupied"
                yield self.env.process(cab.move_to(self.pickup))
                yield self.env.process(cab.move_to(self.dropoff))
                cab.status = "available"
                cab.end_time = self.env.now
                self.wait_time = cab.end_time - self.start_time
                self.real_wait_time = self.wait_time - geodesic(self.pickup, self.dropoff).km / BIKE_SPEED * 60
            else:
                print('>>> FIND_NEAREST_CAB RETURNED None as the output <<<')
                
    def find_nearest_cab(self, cabs):
        nearest_cab = None
        nearest_distance = float("inf")
        for cab in cabs.users:
            if cab.status == "available":
                distance = geodesic(cab.location, self.pickup).km
                if distance < nearest_distance:
                    nearest_cab = cab
                    nearest_distance = distance
        return nearest_cab
    
# Setup and start the simulation
def setup(env, bookings, cabs):
    # Create cabs
    for i in range(NUM_CABS):
        cab = Cab(env, i)
        cabs.put(cab)

    # Create customers
    for id, booking in enumerate(bookings):
        arrival_time, pickup, dropoff = booking
        customer = Customer(env, id, pickup, dropoff)
        env.process(customer.request_cab(cabs, arrival_time))

# Run the simulation
env = simpy.Environment()
cabs = simpy.Resource(env, capacity=NUM_CABS)
env.process(setup(env, BOOKINGS, cabs))
env.run(until=SIM_TIME)

# Print results
for id, booking in enumerate(BOOKINGS):
    # arrival_time, pickup, dropoff = booking
    customer = next((c for c in Customer if c.id == id), None)
    if customer is not None:
        print(f"Customer {id} waited {customer.wait_time:.2f} minutes for a cab.")
    else:
        print(f"Customer {id} did not get a cab.")
```
